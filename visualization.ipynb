```python
import matplotlib.pyplot as plt
file = open('Dataset/reduced_possible_words.txt', 'r')
# Create alphabet
alphabet = []
c = "a"
for i in range(26):
    alphabet.append(c)
    c = chr(ord(c) + 1)

# Create dictionary
alph_dict = {letter: 0 for letter in alphabet}
for word in file:
    for letter in word:
        if letter != '\n':
            alph_dict[letter] += 1

letters = list(alph_dict.keys())
vals = list(alph_dict.values())
plt.title('Letter Frequency in Reduced Dataset')
plt.ylabel('Occurences')
plt.xlabel('Letter')
plt.bar(letters, vals)
plt.show()
file.close()
```

```python
# Compare vowel, consonant appearances in words
import matplotlib.pyplot as plt
file = open('Dataset/reduced_possible_words.txt', 'r')
choices = ['vowel', 'consonant']
vowels = ['a', 'e', 'i', 'o', 'u']

count = {entry: 0 for entry in choices}
for word in file:
    for letter in word:
        if letter in vowels:
            count['vowel'] += 1
        elif letter != '\n':
            count['consonant'] += 1

types = list(count.keys())
vals = list(count.values())
plt.title('Letter Type Frequency in Reduced Dataset')
plt.ylabel('Occurences')
plt.xlabel('Type')
plt.bar(types, vals)
file.close()
```

```python
# Pie chart: appearances of vowels
import matplotlib.pyplot as plt
import numpy as np
file = open('Dataset/reduced_possible_words.txt', 'r')
vowels = ['a', 'e', 'i', 'o', 'u']

vowel_arr = [0, 0, 0, 0, 0]
for word in file:
    word = word[:-1]
    count = 0
    for letter in word:
        if letter in vowels:
            count += 1
    vowel_arr[count] += 1

print(vowel_arr)
plt.pie(vowel_arr, labels = np.arange(0, 5))
plt.show()
file.close()
```

```python
# Find all 2-vowel words
file = open('Dataset/reduced_possible_words.txt', 'r')

# IF 2 vowels exactly, add to list.
vowels = ['a', 'e', 'i', 'o', 'u']
words = []
for word in file:
    word = word[:-1] # Remove newline
    count = 0
    for letter in word:
        if letter in vowels:
            count += 1
    if count == 1:
        words.append(word)
    elif count == 2:
        words.append(word)

print(len(words))
print(words)
file.close()
```
