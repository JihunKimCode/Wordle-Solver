```python
import matplotlib.pyplot as plt
file = open('Dataset/reduced_possible_words.txt', 'r')
# Create alphabet
alphabet = []
c = "a"
for i in range(26):
    alphabet.append(c)
    c = chr(ord(c) + 1)

# Create dictionary
alph_dict = {letter: 0 for letter in alphabet}
for word in file:
    for letter in word:
        if letter != '\n':
            alph_dict[letter] += 1

letters = list(alph_dict.keys())
vals = list(alph_dict.values())
plt.title('Letter Frequency in Reduced Dataset')
plt.ylabel('Occurences')
plt.xlabel('Letter')
plt.bar(letters, vals)
plt.show()
file.close()
```

```python
# Compare vowel, consonant appearances in words
import matplotlib.pyplot as plt
file = open('Dataset/reduced_possible_words.txt', 'r')
choices = ['vowel', 'consonant']
vowels = ['a', 'e', 'i', 'o', 'u']

count = {entry: 0 for entry in choices}
for word in file:
    for letter in word:
        if letter in vowels:
            count['vowel'] += 1
        elif letter != '\n':
            count['consonant'] += 1

types = list(count.keys())
vals = list(count.values())
plt.title('Letter Type Frequency in Reduced Dataset')
plt.ylabel('Occurences')
plt.xlabel('Type')
plt.bar(types, vals)
file.close()
```

```python
# Pie chart: appearances of vowels
import matplotlib.pyplot as plt
import numpy as np
file = open('Dataset/reduced_possible_words.txt', 'r')
vowels = ['a', 'e', 'i', 'o', 'u']

vowel_arr = [0, 0, 0, 0, 0, 0]
for word in file:
    word = word[:-1]
    count = 0
    for letter in word:
        if letter in vowels:
            count += 1
    vowel_arr[count] += 1

print(vowel_arr)
plt.title('Number of vowels in word')
plt.pie(vowel_arr, labels = np.arange(0, 6))
plt.show()
file.close()
```

```python
# Create heatmap of letters for each word position
import matplotlib.pyplot as plt
import seaborn as sns
file = open('Dataset/reduced_possible_words.txt', 'r')
# Create alphabet
alphabet = []
c = "a"
for i in range(26):
    alphabet.append(c)
    c = chr(ord(c) + 1)

first = {letter: 0 for letter in alphabet}
second = {letter: 0 for letter in alphabet}
third = {letter: 0 for letter in alphabet}
fourth = {letter: 0 for letter in alphabet}
fifth = {letter: 0 for letter in alphabet}

# Create dictionary
for word in file:
    first[word[0]] += 1
    second[word[1]] += 1
    third[word[2]] += 1
    fourth[word[3]] += 1
    fifth[word[4]] += 1

# Heatmap array
data = []
for i in [first, second, third, fourth, fifth]:
    data.append(list(i.values()))

# Plot
plt.figure(figsize=(13, 2.5))
plt.title('Letter Frequency Heatmap')
sns.heatmap(data, xticklabels=alphabet, annot=True, fmt='g', cmap='viridis')
plt.show()
file.close()
```
